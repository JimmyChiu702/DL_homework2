# Lab2 - EEG Classification

```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
%matplotlib inline

# Device configuration
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
def read_bci_data():
    S4b_train = np.load('S4b_train.npz')
    X11b_train = np.load('X11b_train.npz')
    S4b_test = np.load('S4b_test.npz')
    X11b_test = np.load('X11b_test.npz')

    train_data = np.concatenate((S4b_train['signal'], X11b_train['signal']), axis=0)
    train_label = np.concatenate((S4b_train['label'], X11b_train['label']), axis=0)
    test_data = np.concatenate((S4b_test['signal'], X11b_test['signal']), axis=0)
    test_label = np.concatenate((S4b_test['label'], X11b_test['label']), axis=0)

    train_label = train_label - 1
    test_label = test_label - 1
    train_data = np.transpose(np.expand_dims(train_data, axis=1), (0, 1, 3, 2))
    test_data = np.transpose(np.expand_dims(test_data, axis=1), (0, 1, 3, 2))

    mask = np.where(np.isnan(train_data))
    train_data[mask] = np.nanmean(train_data)

    mask = np.where(np.isnan(test_data))
    test_data[mask] = np.nanmean(test_data)

    train_data = torch.Tensor(train_data)
    train_label = torch.Tensor(train_label).long()
    test_data = torch.Tensor(test_data)
    test_label = torch.Tensor(test_label).long()
    
    print(train_data.shape, train_label.shape, test_data.shape, test_label.shape)

    return train_data, train_label, test_data, test_label

trainX, trainY, testX, testY = read_bci_data()
```

```python
def data_loader(x, bs=64):
    for i in range(len(x) // bs):
        yield x[i*bs:(i+1)*bs]
    if len(x)%bs != 0:
        yield x[(i+1)*bs:]

def runModel(model, 
             trainX, trainY, testX, testY, 
             batch_size=64,
             learning_rate=1e-02,
             num_epochs=300,
             optimizer=torch.optim.Adam,
             loss_func=nn.CrossEntropyLoss()
            ):
    
    # Print the model info
    # print(model)
    
    # Loss and optimizer
    criterion = loss_func
    optimizer = optimizer(model.parameters(), lr=learning_rate)
    
    # Accuracy list, including training amd testing
    acc_lists = {'train': [], 'test': []}
    
    # Train & test
    for epoch in range(num_epochs):
    
        # Train
        model.train()
    
        correct, total = 0, 0
        idx = torch.randperm(trainY.size(0))
        for x, y in zip(data_loader(trainX[idx], batch_size), data_loader(trainY[idx], batch_size)):
            x = x.to(device)
            y = y.to(device)
            
            # Forward
            outputs = model(x)
            loss = criterion(outputs, y)
            _, preds = torch.max(outputs.data, 1)
            correct += (preds==y).sum().item()
            total += y.size(0)
            
            # Back propagation & optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        acc_lists['train'].append((correct/total)*100)
        
        # Test
        model.eval()
        
        correct, total = 0, 0
        with torch.no_grad():
            for x, y in zip(data_loader(testX[idx], batch_size), data_loader(testY[idx], batch_size)):
                x = x.to(device)
                y = y.to(device)
                
                # Forward
                outputs = model(x)
                _, preds = torch.max(outputs.data, 1)
                correct += (preds==y).sum().item()
                total += y.size(0)
        acc_lists['test'].append((correct/total)*100)
    
    return acc_lists
```

## EEGNet

```python
class EEGNet(nn.Module):
    def __init__(self, activation_func='ReLU', dropout=0.25):
        super(EEGNet, self).__init__()
        
        if activation_func == 'ReLU':
            activation = nn.ReLU()
        elif activation_func == 'LeakyReLU':
            activation = nn.LeakyReLU()
        elif activation_func == 'ELU':
            activation = nn.ELU()
        else:
            raise KeyError('Activation function should be either "ReLU", "LeakyReLU", or "ELU"')
            
        self.firstConv = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=(1, 51), stride=(1, 1), padding=(0, 25), bias=False),
            nn.BatchNorm2d(16, eps=1e-05, momentum=0.1, affine=True)
        )
        self.depthwiseConv = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=(2, 1), stride=(1, 1), groups=16, bias=False),
            nn.BatchNorm2d(32, eps=1e-05, momentum=0.1, affine=True),
            activation,
            nn.AvgPool2d(kernel_size=(1, 4), stride=(1, 4), padding=0),
            nn.Dropout(p=dropout)
        )
        self.separableConv = nn.Sequential(
            nn.Conv2d(32, 32, kernel_size=(1, 15), stride=(1, 1), padding=(0, 7), bias=False),
            nn.BatchNorm2d(32, eps=1e-05, momentum=0.1, affine=True),
            activation,
            nn.AvgPool2d(kernel_size=(1, 8), stride=(1, 8), padding=0),
            nn.Dropout(p=dropout)
        )
        self.classifier = nn.Sequential(
            nn.Linear(736, 2)
        )
    
    def forward(self, x):
        x = self.firstConv(x)
        x = self.depthwiseConv(x)
        x = self.separableConv(x)
        x = torch.flatten(x, start_dim=1)
        x = self.classifier(x)
        return x
```

## DeepConvNet

```python
class DeepConvNet(nn.Module):
    def __init__(self, activation_func='ReLU'):
        super(DeepConvNet, self).__init__()
        
        if activation_func == 'ReLU':
            activation = nn.ReLU()
        elif activation_func == 'LeakyReLU':
            activation = nn.LeakyReLU()
        elif activation_func == 'ELU':
            activation = nn.ELU()
        else:
            raise KeyError('Activation function should be either "ReLU", "LeakyReLU", or "ELU"')
            
        self.firstConv = nn.Sequential(
            nn.Conv2d(1, 25, kernel_size=(1, 5)),
            nn.Conv2d(25, 25, kernel_size=(2, 1)),
            nn.BatchNorm2d(25, eps=1e-05, momentum=0.1),
            activation,
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(p=0.5)
        )
        self.secondConv = nn.Sequential(
            nn.Conv2d(25, 50, kernel_size=(1, 5)),
            nn.BatchNorm2d(50, eps=1e-05, momentum=0.1),
            activation,
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(p=0.5)
        )
        self.thirdConv = nn.Sequential(
            nn.Conv2d(50, 100, kernel_size=(1, 5)),
            nn.BatchNorm2d(100, eps=1e-05, momentum=0.1),
            activation,
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(p=0.5)
        )
        self.fourthConv = nn.Sequential(
            nn.Conv2d(100, 200, kernel_size=(1, 5)),
            nn.BatchNorm2d(200, eps=1e-05, momentum=0.1),
            activation,
            nn.MaxPool2d(kernel_size=(1, 2)),
            nn.Dropout(p=0.5)
        )
        self.classifier = nn.Sequential(
            nn.Linear(8600, 2)
        )
    
    def forward(self, x):
        x = self.firstConv(x)
        x = self.secondConv(x)
        x = self.thirdConv(x)
        x = self.fourthConv(x)
        x = torch.flatten(x, start_dim=1)
        x = self.classifier(x)
        return x
```

## Experiments

```python
# Experiment reproducibility
seed = 95
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.cuda.manual_seed_all(seed)
np.random.seed(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# Hyper parameters
batch_size = 64
learning_rate = 1e-2
num_epochs = 300

activations = ['ReLU', 'LeakyReLU', 'ELU']
ACC = {net: {act: None for act in activations} for net in ['EEGNet', 'DeepConvNet']}

# EEGNet
print('EEGNet')
EEGNet_acc = {act: None for act in activations}
for activation in activations:
    model = EEGNet().to(device)
    acc = runModel(model, trainX, trainY, testX, testY)
    EEGNet_acc[activation] = acc
    print('{} - Training accuracy: {}, Testing accuracy: {}'.format(activation, acc['train'][-1], acc['test'][-1]))
print()
    
# DeepConvNet
print('DeepConvNet')
DeepConvNet_acc = {act: None for act in activations}
for activation in activations:
    model = DeepConvNet().to(device)
    acc = runModel(model, trainX, trainY, testX, testY)
    DeepConvNet_acc[activation] = acc
    print('{} - Training accuracy: {}, Testing accuracy: {}'.format(activation, acc['train'][-1], acc['test'][-1]))
```

```python
# Result comparison
fig = plt.figure(figsize=(15, 10))
plt.title('Activation function comparison (EEGNet)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')

legend_labels = []
for activation in activations:
    for mode in ['train', 'test']:
        plt.plot(EEGNet_acc[activation][mode])
        legend_labels.append('{}_{}'.format(activation, mode))
plt.legend(legend_labels, loc='lower right')
```
